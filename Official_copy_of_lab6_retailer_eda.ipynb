## Create Basic Charts

➡️ Assignment Tasks
Your target variable is Total Sales (count of sales)
- Create a bar chart
- Create a histogram
- Create a box plot
- Create a scatter plot
Each visualization should bring some insight into the total sales.

```python
#bar chart

import matplotlib.pyplot as plt

monthly_sales = df.groupby('order_month')['total_sales'].sum()

plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='lightblue')

plt.xlabel('Order Month')
plt.ylabel('Total Sales Revenue')
plt.title('Total Sales Revenue per Month')

plt.xticks(rotation=45, ha='right')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()
```

```python
#histogram

import matplotlib.pyplot as plt

monthly_sales = df.groupby('order_month')['total_sales'].sum()

plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='orange')

plt.xlabel('Order Month')
plt.ylabel('Total Sales Revenue')
plt.title('Total Sales Revenue per Month')

plt.xticks(rotation=45, ha='right')
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.tight_layout()
plt.show()

```

```python
#box plot

plt.figure(figsize=(8, 6))
plt.boxplot(df['total_sales'])

plt.xlabel('Total Sales Revenue')
plt.title('Distribution of Total Sales Revenue')

plt.tight_layout()
plt.show()
```

## Explore Relationships

➡️ Assignment Tasks
Your target variable is Total Sales (count of sales)
- How does the month of year affect sales?
- What are the best states for sales?
- What are the best and worst products for sales?

```python
#month of year

import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(df)

# Extract month from order_year_month and create a new column
df['order_month'] = pd.to_datetime(df['order_year_month']).dt.month

# Group by month and calculate total sales for each month
monthly_sales = df.groupby('order_month')['total_sales'].sum()

# Plot the sales data
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales, marker='o', color='skyblue', linestyle='-')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

```python
#sales by state

state_sales = df.groupby('state_province')['total_sales'].sum()
best_states = state_sales.nlargest(10)

print("Highest Sales by State")
print(best_states)
```

```python
#sales by product

product_sales = df.groupby('product_name')['total_sales'].sum()
best_product = product_sales.idxmax()
worst_product = product_sales.idxmin()

print("Highest Selling Product:", best_product)
print("Lowest Selling Product:", worst_product)
```
