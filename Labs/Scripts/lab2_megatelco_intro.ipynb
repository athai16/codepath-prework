## Clean up the data
Did you notice anything unusual about the "house" amounts?

How about the handset price and income?

```python
df_clean = df[(df['house'] > 0) & (df['income'] > 0) & (df['handset_price'] < 1000)]
df_clean.describe()
```

# Add Derived Attribute

Add a column with the income in K (thousands)

```python
df_clean['income_k'] = (df_clean['income']/1000).round(2)
df_clean.head()
```

## Summarize the Data
 What is average house price at the two levels of leave?

 Conclusion?  Stayers have substantially more expensive homes.

```python
df_clean[['house','leave']].groupby('leave').mean().round(2)
```

## Additional Data Preparation

➡️ Assignment Tasks
- Add at least two additional steps to clean or filter the dataset

```python
# change college column so that 'one' = 'yes' and 'zero = 'no'

df_clean['college'] = df_clean['college'].replace({'one': 'yes', 'zero': 'no'})

print(df_clean)
```

```python
# filter rows to show customers who plan to leave the phone plan

df_leave = df_clean[df_clean['leave'].str.contains('LEAVE', case=False)]
print(df_leave)
```

## Data Visualization

➡️ Assignment Tasks
- Add at least two charts to show a summary of any major trends

```python
# scatter plot to show the relationship between income and handset price

plt.scatter(df_clean['income'], df_clean['handset_price'])
plt.title('Income and Handset Price')
plt.xlabel('Income')
plt.ylabel('Handset Price')
plt.show()
```

```python
# pie chart that shows the distribution of reported usage level

import matplotlib.pyplot as plt

plt.figure(figsize=(4, 4))
df_clean['reported_usage_level'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title('Distribution of Reported Usage Level')
plt.ylabel('')
plt.show()
```
